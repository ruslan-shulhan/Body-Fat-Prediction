```python
# Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
```

<h3>Data Collection:</h3>

```python
# Loading "raw" data to pandas' DataFrame for file "bodyfat_1.csv" data.
df_1 = pd.read_csv('../../Data/bodyfat_1.csv')
print(df_1.head())
```

```python
df_2 = pd.read_csv('../../Data/bodyfat_2.csv')
print(df_2.head())
```

```python
# Data joining
df_1.info()
```

```python
df_2.info()
```

```python
df = pd.concat([df_1, df_2], axis=1)
df.head()
```

<h3>Data Organization:</h3>

```python
# Changing column's names
col_names = ['density', 'bodyfat', 'age', 'weight', 'height', 'neck', 'chest', 'abdomen', 'hip', 'thigh', 'knee', 'ankle', 'biceps', 'forearm', 'wrist']
df.columns = col_names
df.head()
```

```python
# Checking for missing values and dTypes of all columns
df.info()
```

```python
# Converting some columns to a necessary dTypes
df = df.apply(pd.to_numeric, errors='coerce')
df.info()
```

```python
df.describe()
```

```python
# Drop zero values with NaNs
df.isnull().values.any()
```

```python
df.isnull().sum()
```

```python
df[df.isna().any(axis=1)]
```

```python
df.dropna(axis=0, how='any', inplace=True)
df.isnull().sum()
```

```python
df.describe()
```

```python
df[df['hip'] == 0]
```

```python
df.drop(labels=[253], axis=0, inplace=True)
df.describe()
```

<h3>Data definition:</h3>

```python
# Column names
df.columns
```

```python
# dTypes
df.dtypes
```

```python
# Convert column "weight" from lbs to kgs
df['weight'] = np.around((df['weight'] / 2.205), decimals=1)
df['weight']
```

```python
# Convert column "height" from inch to cm
df['height'] = np.around((df['height'] * 2.54), decimals=1)
df['height']
```

```python
# Unique values by percentage
def unique_values_fun(col_names, dataFrame):
    """
    Counts the unique values in the provided column names.
        col_names: List -> The column names.
        dataFrame: pd.DataFrame -> The data to look at.
        return: dictionary -> The return product is dictionary with recored unque values.
    """
    dic_result = {}
    if len(col_names) == 0 or df.empty:
        return dic_result    
    
    for col_name in col_names:
        if col_name in dic_result.keys():
            dic_result[col_name + "_%"] += np.around(((len(pd.unique(dataFrame[col_name])) / len(dataFrame[col_name])) * 100), decimals=0)
        else:
            dic_result[col_name + "_%"] = np.around(((len(pd.unique(dataFrame[col_name])) / len(dataFrame[col_name])) * 100), decimals=0)
            
    return dict(sorted(dic_result.items(), key=lambda item: item[1]))
```

```python
uniq_values = pd.DataFrame(unique_values_fun(col_names=col_names, dataFrame=df), index=[0])
uniq_values
```

```python
# Range of values
def range_values_fun(col_names, dataFrame):
    """
    Counts the range of values in the provided column names.
        col_names: List -> The column names.
        dataFrame: pd.DataFrame -> The data to look at.
        return: dictionary -> The return product is dictionary with recored range of values.
    """
    dic_result = {}
    if len(col_names) == 0 or df.empty:
        return dic_result    
    
    for col_name in col_names:
        if col_name in dic_result.keys():
            dic_result[col_name] += np.around(np.amax(dataFrame[col_name]) - np.min(dataFrame[col_name]), decimals=2)
        else:
            dic_result[col_name] = np.around(np.amax(dataFrame[col_name]) - np.min(dataFrame[col_name]), decimals=2)
            
    return dict(sorted(dic_result.items(), key=lambda item: item[1]))
```

```python
range_of_values = pd.DataFrame(range_values_fun(col_names=col_names, dataFrame=df), index=[0])
range_of_values
```
