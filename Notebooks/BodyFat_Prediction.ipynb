```python
# Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
```

<h3>Data Collection:</h3>

```python
# Loading "raw" data to pandas' DataFrame for file "bodyfat_1.csv" data.
df_1 = pd.read_csv('../Data/bodyfat_1.csv')
print(df_1.head())
```

```python
df_2 = pd.read_csv('../Data/bodyfat_2.csv')
print(df_2.head())
```

```python
# Data joining
df_1.info()
```

```python
df_2.info()
```

```python
df = pd.concat([df_1, df_2], axis=1)
df.head()
```

<h3>Data Organization:</h3>

```python
# Changing column's names
col_names = ['density', 'bodyfat', 'age', 'weight', 'height', 'neck', 'chest', 'abdomen', 'hip', 'thigh', 'knee', 'ankle', 'biceps', 'forearm', 'wrist']
df.columns = col_names
df.head()
```

```python
# Checking for missing values and dTypes of all columns
df.info()
```

```python
# Converting some columns to a necessary dTypes
df = df.apply(pd.to_numeric, errors='coerce')
df.info()
```

```python
df.describe()
```

```python
# Drop zero values with NaNs
df.isnull().values.any()
```

```python
df.isnull().sum()
```

```python
df[df.isna().any(axis=1)]
```

```python
df.dropna(axis=0, how='any', inplace=True)
df.isnull().sum()
```

```python
df.describe()
```

```python
df[df['hip'] == 0]
```
